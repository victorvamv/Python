```python
#ESTRUCTURAL
num1=0
num2=0

num1 = int(input('Ingresa el primer número: '))
num2 = int(input('Ingresa el segundo número: '))

while True:
  opcion=int(input('''
  Selecciona una opcion
  1 - SUMAR
  2 - RESTAR
  3 - DIVIDIR
  4 - MULTIPLICAR
  5 - SALIR'''))

  if(opcion == 1):
    res = num1 + num2
    print(f'La suma de {num1} más {num2} es {res} ')

  elif(opcion == 2):
    res = num1 - num2
    print(f'La resta de {num1} menos {num2} es {res} ')

  elif(opcion == 3):
    if((num1 != 0) and (num2 != 0) ):
      res = num1/num2
      print(f'La division de {num1} entre {num2} es {res} ')
    else:
      print(f'División de 0 ')

  elif(opcion == 4):
    res = num1 * num2
    print(f'El producto de {num1} por {num2} es {res} ')

  elif(opcion == 5):
    break

  else:
    print(f'Opción no valida ')
```
